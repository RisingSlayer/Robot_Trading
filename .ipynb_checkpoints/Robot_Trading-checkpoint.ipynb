```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
import yfinance as yf
import datetime
import time
from bs4 import BeautifulSoup
from IPython.display import clear_output

global df_bitcoin, precio_actual, tendencia, media_bitcoin, algoritmo_decision
```

# 2. Obtención de datos

## 2.1 Precios Históricos del Bitcoin

```python
def importar_base_bitcoin():
    global df_bitcoin, precio_actual, tendencia, media_bitcoin,  algoritmo_decision
    bitcoin_data = yf.download(tickers ='BTC-USD', period ='7d', interval ='5m')

    df_bitcoin = pd.DataFrame(bitcoin_data)
```

```python
importar_base_bitcoin()
df_bitcoin.head()
```

## 2.2 Indicadores de Tendencias

```python
def extraer_tendencias():
    global precio_actual, tendencia

    url = "https://coinmarketcap.com/currencies/bitcoin/"

    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        bitcoin_price_element = soup.find('div', class_ = 'sc-f70bb44c-0 flfGQp flexStart alignBaseline')

        precio_texto = bitcoin_price_element.find('span', class_ = 'sc-f70bb44c-0 jxpCgO base-text').text
        precio_actual = float(precio_texto.replace('$', '').replace(',', ''))

        variation_element = bitcoin_price_element.find('span', class_ = 'sc-f70bb44c-0 jxpCgO base-text').find_next_sibling()

        variacion_texto = variation_element.text.strip('%\xa0(1d)').replace(',', '')
        variacion = float(variacion_texto)

        if variacion < 0:
            tendencia = 'baja'
        else:
            tendencia = 'alta'
    else:
        print("Error al cargar la página de CoinMarketCap")
```

```python
extraer_tendencias()
print('Bitcoin Price: ', precio_actual)
print('Tendencia: ', tendencia)
```

# 3. Limpieza de datos

```python
print(df_bitcoin.info())
df_bitcoin.describe().round(0)
```

```python
df_bitcoin.boxplot(['Close']);
```

```python
def limpieza_datos():
    global df_bitcoin_limpio, df_bitcoin, precio_actual, tendencia, media_bitcoin, algoritmo_decision

    df_bitcoin_limpio = df_bitcoin.copy()

    df_bitcoin_limpio = df_bitcoin_limpio.drop('Adj Close', axis = 1).reset_index()

    duplicados_en_datetime = df_bitcoin_limpio['Datetime'].duplicated(keep = 'first')

    valores_nulos = df_bitcoin_limpio["Close"].isnull().sum()
    
    if valores_nulos > 0:
        df_bitcoin_limpio["Close"].fillna(df_bitcoin_limpio["Close"].mean(), inplace=True)

    df_bitcoin_limpio = df_bitcoin_limpio[df_bitcoin_limpio["Volume"] > 0]

    registros_eliminados = len(df_bitcoin) - len(df_bitcoin_limpio)

    Q1 = df_bitcoin_limpio["Close"].quantile(0.25)
    Q3 = df_bitcoin_limpio["Close"].quantile(0.75)

    df_bitcoin_limpio = df_bitcoin_limpio[(df_bitcoin_limpio["Close"] >= Q1) & (df_bitcoin_limpio["Close"] <= Q3)]

    registros_eliminados = len(df_bitcoin) - len(df_bitcoin_limpio)
    media_bitcoin = df_bitcoin_limpio["Close"].mean()
```

```python
limpieza_datos()
```

# 4. Toma de decisiones

```python
def tomar_decisiones():
    global df_bitcoin_limpio, df_bitcoin, precio_actual, tendencia, media_bitcoin, algoritmo_decision

    if (precio_actual >= media_bitcoin) & (tendencia == 'baja'):
        algoritmo_decision = 'Vender'
    elif (precio_actual < media_bitcoin) & (tendencia == 'alta'):
        algoritmo_decision = 'Comprar'
    else:
        algoritmo_decision = 'Esperar'
```

```python
tomar_decisiones()
print('Precio Promedio Bitcoin:', media_bitcoin)
print('Precio Actual Bitcoin:', precio_actual)
print('Tendencia:', tendencia)
print('Decisión del Algoritmo:', algoritmo_decision)
```

# 5. Visualización

```python
def visualizacion():
    global df_bitcoin_limpio, df_bitcoin, precio_actual, tendencia, media_bitcoin, algoritmo_decision
    
    df_bitcoin[['Promedio']] = media_bitcoin
    plt.figure(figsize = (16, 5))
    plt.title('Precio de Bitcoin ($)', fontsize = 30, fontweight = 'bold', color = 'black', loc = 'left')
    df_bitcoin['Close'].plot()
    df_bitcoin['Promedio'].plot()
    plt.xlabel('')
    plt.xticks(rotation = 0)
    
    if algoritmo_decision == 'Vender':
        color_decision = 'green'
    elif algoritmo_decision =='Comprar':
        color_decision = 'red'
    else:
        color_decision = 'orange'

    plt.annotate(algoritmo_decision,
                 xy = (df_bitcoin.index[-1], df_bitcoin['Close'].iloc[-1]),
                 xytext = (-10, 65),
                 textcoords = 'offset points',
                 ha = 'center', va = 'bottom', fontsize = 20, fontweight = 'bold', color = color_decision)

    fecha_hora_actual = datetime.datetime.now()

    print('Hora de ejecución:', fecha_hora_actual)
    print('Precio Promedio Bitcoin:', media_bitcoin)
    print('Precio Actual Bitcoin:', precio_actual)
    print('Tendencia:', tendencia)
    plt.show()
```

```python
visualizacion()
```

# 6. Automatización

```python
while(True):
    clear_output()
    importar_base_bitcoin()
    extraer_tendencias()
    limpieza_datos()
    tomar_decisiones()
    visualizacion()
    time.sleep(300)
```
